```python
# 1.1 Basic feature extraction using textdata

import requests
#requests permissions to get the content
r = requests.get("https://github.com/unt-iialab/info5731_spring2021/blob/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt")
text = r.text
```

```python
#importing natural language toolkit for our text operations
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import re
```

```python
# Number of sentences
#splitting sentences using sent_tokenize
sentences = nltk.sent_tokenize(text)
print("Number of sentences:", len(sentences))

# Number of words
# We use split command to split sentence into words and then count number of words
number_of_words = []
for sent in sentences:
    words = re.split(r'\W+',sent)
    for i in words:
        number_of_words.append(i)
print("Number of words:", len(number_of_words))

# Number of characters
# we are using a for loop to get the characters after splitting words
characters = 0
for i in number_of_words:
  characters+=len(i)
print("Number of characters:", characters)

# Average word length
# calculating the average word length as total characters by number of words

print("Average word length:", characters/len(number_of_words))

# Number of stopwords
# First we are getting stopwords of english from lib and then comparing our words and counting if it is a stopword

from nltk.corpus import stopwords
stop_words = [word for word in number_of_words if word in stopwords.words('english')]
print("Number of stopwords",len(stop_words))

# Number of numerics
# we are checking for digits and counting them using for loop
numeric =(map(str.isdigit, re.split(r'\W+',text)))
sum=0 
for i in numeric:
  sum+=i
print("Number of numerics:",sum)

# Number of uppercase words
# we are looking for upper case words and counting them using for loop same as above
uppercase =(map(str.isupper, re.split(r'\W+',text))) 
sum=0 
for i in uppercase:
  sum+=i
print("Number of uppercase words:",sum)
        
```

```python
# stemming
# we are using porterstemmer from nltk library for this method, which removes the suffices

from nltk.stem import PorterStemmer 
ps = PorterStemmer() 
for w in number_of_words: 
    print(w, " : ", ps.stem(w)) 
```

```python
# Lemmatization
# we are using worldnetlemmatizer method to obtain the root word

from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
for i in number_of_words:
  print(lemmatizer.lemmatize(i))
```
